```python
# 필요한 모듈들 import
from os.path import join
from glob import glob
import hashlib
```

```python
# 구글 드라이브 마운트
from google.colab import drive
drive.mount('/content/drive')

# 파일 디렉토리의 경로
data_path = '/content/drive/MyDrive/Colab Notebooks/network'

# 경로에 있는 txt파일 목록 리스트에 저장
data_list = sorted(glob(join(data_path, '*.txt')))
data_list
```

```python
# 주어진 패스워드 테이블로 해시 딕셔너리 생성
word_list = open(data_list[1]) # password table

hash_dic = dict()
for word in word_list:
  word = word.strip()
  hashed_word = hashlib.md5(word.encode('utf-8')).hexdigest()
  hash_dic[hashed_word]=word # 해시값을 키로 설정
hash_dic
hash_dic['363b122c528f54df4a0446b6bab05515']='j'
```

```python
# 위에서 만든 해시테이블과 과제로 받은 hash값 비교
md5_list = open(data_list[0]) # hash table

cnt=0
for md5 in md5_list:
  md5 = md5.strip()
  if md5 in hash_dic:
    cnt = cnt+1
    print(cnt,'/1000000 password has been cracked, hashed:', md5 ,'cracked:',hash_dic[md5])
    del hash_dic[md5]
  else:
    print(md5)
  
```
